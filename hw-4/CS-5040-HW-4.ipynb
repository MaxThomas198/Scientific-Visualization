Part 1 (Volume Rendering):

1.) Indirect volume renderings are created from a deriveded data set that is then rendered to created an approximate volume rendering of a data set. A direct volume rendering directly renders the actual volume of a data set without altering data. One example of direct volume rendering is raycasting. A volume rendering of an isosurface is an indirect volume rendering. Basically anytime you create a mesh rendering that process will be indirect.

2.) The raycasting composite operation helps determine what color a point should be. It looks at the a ray of ordered points and calulates based on a transfer function and how transparent a point is, what color should appear on the screen. 

3.) A first threshold would produce a result that looks like an iso surface. It does this because each ray will stop at the first point it reaches in the data set which displays a volume rendering that looks pretty similar to an iso surface.

4.) One reason Slice based rendering is much faster than Raycasting is because you don't necessarily have to render each slice to create the image that you want. Raycasting also takes a very long time because you need to calculate values for each indivdual point unlike in slice rendering where you can approimate voxels in order to save time.

5.) The first step in the marching cube algorithm is to calculate 8 boolean values to represent a cube and then those values are turned into an 8 bit bitfield with 256 possible values. The next step is to generate vertices for the 12 edges of the cube and create a look-up tabel for each edge indices for the vertexs. To get a more accurate final mesh we can interpolate the vertex position on the edge based on difference between sample values of the cube corners.
